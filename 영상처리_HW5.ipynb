```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
BGR_img = cv2.imread('/content/drive/MyDrive/ImageProcessingAssignment/rena-s.jpg')
```

```python
white = np.array([255, 255, 255], np.uint8)
```

```python
CMY_img = white - BGR_img
```

```python
yellow = CMY_img[:,:,0].copy()
magenta = CMY_img[:,:,1].copy()
cyan = CMY_img[:,:,2].copy()
```

```python
print("BGR.color")
cv2.putText(BGR_img, "BGR.color", (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 1, (180,105, 255), 1, cv2.LINE_AA)
cv2_imshow(BGR_img)
#ㅡㅡ
print("CMY.color")
cv2.putText(CMY_img, "CMY.color", (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 1, (180,105, 255), 1, cv2.LINE_AA)
cv2_imshow(CMY_img)
#ㅡㅡ
print("yellow.color")
cv2.putText(yellow, "yellow.color", (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 0.7, 0, 1, cv2.LINE_AA)
cv2_imshow(yellow)
#ㅡㅡ
print("magenta.color")
cv2.putText(magenta, "magenta.color", (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 0.7, 0, 1, cv2.LINE_AA)
cv2_imshow(magenta)
#ㅡㅡ
print("cyan.color")
cv2.putText(cyan, "cyan.color", (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 0.7, 0, 1, cv2.LINE_AA)
cv2_imshow(cyan)
```
