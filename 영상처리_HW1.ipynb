```python
import cv2
print(cv2.__version__)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
title1 = 'gray2gray'
title2 = 'gray2color'
```

```python
from google.colab.patches import cv2_imshow

gray2gray = cv2.imread("/content/drive/MyDrive/ImageProcessingAssignment/rena-s.jpg", cv2.IMREAD_GRAYSCALE) #imread 메소드는 파일에 저장된 이미지를 읽어서 Numpy 배열 형태로 반환한다.
gray2color = cv2.imread("/content/drive/MyDrive/ImageProcessingAssignment/rena-s.jpg", cv2.IMREAD_COLOR) #cv2.~ 이건 옵션 상수

if gray2gray is None or gray2color is None :
  raise Exception("Image File Read Error !!")

print("pixel information [100, 100]")
print("%s %s" % (title1, gray2gray[100, 100])) #100, 100은 픽셀값, (100,100) 위치에 있는 픽셀값을 꺼내는 것
print("%s %s" % (title2, gray2color[100, 100]))

print('[gray2gray]') #title 대체 print
cv2_imshow(gray2gray) #코랩에서는 cv2.imshow(title1, gray2gray) 막혀있음.. 로컬 GUI창을 띄우는 함수인데 창을 열 수 없기 때문에
print('[gray2color]')
cv2_imshow(gray2color) #cv2.imshow나 cv2_imshow나 둘다 Numpy 배열을 화면에서 시각화해서 보여주는 함수이다, 배열을 이미지처럼 렌더링해주는 함수.
#렌더링(Rendering) = 데이터를 사람이 볼 수 있는 형태로 변환해서 화면에 그리는 과정
#cv2.waitKey(0)    # 코랩에서 무의미
#cv2.waitKey(0)    # 키가 눌릴 때까지 무한 대기
#cv2.waitKey(1000) # 1000ms = 1초 대기 or 키가 눌릴때까지, 1초 안에 눌리면 키 코드 반환, 1초 동안 아무 키도 안 누르면 -1 반환..
```
