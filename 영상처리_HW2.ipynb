```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/dr33ive/MyDrive/ImageProcessingAssignment/rena-s.jpg")

if image is None:
  print("Could not open or find the image")
  exit(0)
```

```python
new_image = np.zeros(image.shape, image.dtype) # image와 동일한 크기와 자료형을 가진 0으로 채워진 배열을 만든다.
print(image.dtype) #0~255
alpha = 1.5 #Simple contrast control
beta = 50 #Simple brightness control

for y in range(image.shape[0]):
  for x in range(image.shape[1]):
    for c in range(image.shape[2]):
      new_image[y, x, c] = np.clip(alpha * image[y, x, c] + beta, 0, 255) #np.clip(배열, min, max) min보다 작으면 min으로 max보다 크면 max로 저장
```

```python
originName = "origin"
destName = "target"

# blue = (255, 0, 0) # green = (0, 255, 0) # red = (0, 0, 255)
# custom_color = (40, 90, 255) # BGR순서로 저장된다.

custom_color = (180, 105, 255)

cv2.putText(image, originName, (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 1, custom_color, 1, cv2.LINE_AA) #inplace연산이라 새로 다지우고 쓰려면 이미지 다시 불러와야..
#인자 순서대로, 이미지배열, 출력문자, 시작 좌표, 글꼴, 글자크기, 글자색, 두께, 선 그리는 방식
cv2.putText(new_image, destName, (10, 25), cv2.FONT_HERSHEY_TRIPLEX, 1, custom_color, 1, cv2.LINE_AA)

#cv2.namedWindow(originName)
#cv2.namedWindow(destName) 코랩 환경이라 안돌아가 주석처리

#cv2.moveWindow(originName, 30, 30)
print(originName) #코랩이라 imshow(originName, image) 대체하기
cv2_imshow(image)

#cv2.moveWindow(destName, 250, 30)
print(destName) #코랩이라 imshow(destName, image) 대체하기
cv2_imshow(new_image)

#cv2.waitKey(0) 코랩이라 생략했습니다.
#cv2.destroyAllWindows()
```
